```python
#固定的四個套件
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

用下列函式畫出一個彩色螺旋狀的圖

截圖 2025-03-05 20.01.56.png

```python


def spiral_cosine_wave(t, a=1, b=0.1):
    # 用sin和cos和參數 a, b 生成螺旋線
    x = a * np.cos(t) * np.exp(b * t)
    y = a * np.sin(t) * np.exp(b * t)
    return x, y

# 創建角度範圍(0 ~ 10pi)，以 t 表示
t = np.linspace(0, 10 * np.pi, 5000)

#帶入公式
x, y = spiral_cosine_wave(t)

# 畫出螺旋線
plt.figure(figsize=(10, 10)) #設定圖形大小
plt.scatter(x, y, c=x, cmap='inferno', linewidths=1) #調整線的顏色根據x的範圍有所變化，寬度為1
plt.title("Spiral with Cosine Wave Variation")  #標題
plt.axis('equal') #兩個軸的單位長度設為相同，避免變形
plt.colorbar(label='X') #顯示不同Ｘ值對應到的顏色
plt.show() #顯示圖形
```
